```python
class LinkedList:
    def __init__(self):
        self.head = None # linked list is initially empty
    
    def addFirst(self, data): # O(1)
        newNode = Node(data)
        if self.head == None: # if linked list is empty
            self.head = newNode
        else:
            orgHead = self.head
            self.head = newNode
            self.head.next = orgHead
    
    def addIndex(self, data, idx):
        if idx == 0:
            self.addFirst(data)
        else:
            cur = self.head
            i = 0
            while i < idx-1 and cur != None:
                cur = cur.next
                i += 1
            
            if cur != None:   
                newNode = Node(data)
                org_next = cur.next
                cur.next = newNode
                cur.next.next = org_next
            else:
                print("Error!, Index is too large!")
    
    def printLinkedList(self): # O(n)
        if self.head == None:
            print("Linked List is empty!")
        else:
            cur = self.head
            while cur != None:
                print(cur.data, "->", end=" ")
                cur = cur.next
            print("None")
```

```python
a = LinkedList() # initially empty

a.addFirst(40)
a.addFirst(30)
a.addFirst(20)
a.addFirst(10)
a.printLinkedList()
a.addIndex(50, 2)
a.printLinkedList()
a.addIndex(100, 5)
a.printLinkedList()
```

```python
def reverse(head):
    if head == None or head.next == None:
        return head
    prev = None
    cur = head
    while cur != None:
        org_next = cur.next
        cur.next = prev
        prev = cur
        cur = org_next
    return prev
```

```python
a.printLinkedList()
a.head = reverse(a.head)
a.printLinkedList()
```

```python
def checkCyle(head):
    if head == None:
        return False
    slow = head
    fast = head.next
    if fast == None:
        return False
    while slow != fast and fast.next != None and fast.next.next != None:
        slow = slow.next
        fast = fast.next.next
    if slow == fast:
        return True
    else:
        return False
```

```python
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)

n1.next = n2
n2.next = n3
n3.next = n1

checkCyle(n1)
```

```python
def findLastKthElem(head, k):
    if head == None:
        return None
    p1 = head
    p2 = head
    for i in range(k):
        if p1 != None:
            p1 = p1.next
        elif p1 == None:
            return None
    while p1 != None:
        p1 = p1.next
        p2 = p2.next
    return p2.data
```
