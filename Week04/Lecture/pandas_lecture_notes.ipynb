```python
# calculate the average price for all meals
orders["item_price"].str.slice(1).astype(float).mean()
```

```python
orders["item_price"].str.replace("$", "").astype(float).mean()
```

```python
#calculate the average price for the meals whose name is Chicken Bowl
df = orders[orders.item_name == "Chicken Bowl"]
df["item_price"].str.replace("$", "").astype(float).mean()
```

```python
#calculate the average price for the meals whose name contains any form of chicken

orders[orders["item_name"].str.lower().str.contains("chicken")]["item_price"].str.replace("$", "").astype(float).mean()
```

```python
drinks = pd.read_csv("http://bit.ly/drinksbycountry")
drinks.head()
```

```python
# find the country name that consumes the beer the most
drinks.sort_values("beer_servings", ascending=False)["country"].values[0]
```

```python
drinks.sort_values("wine_servings", ascending=False)["country"].values[0]
```

```python
# find the country name in Europe that consumes the beer the most
eu_df = drinks[drinks.continent == "Europe"]
eu_df.sort_values("beer_servings", ascending=False)["country"].values[0]
```

```python
# find the average beer cons in EU
drinks[drinks.continent == "Europe"]["beer_servings"].mean()
```

```python
# find avg beer cons for each continent
continents = drinks["continent"].unique()
for cont in continents:
    res = drinks[drinks.continent == cont]["beer_servings"].mean()
    print(cont, res)
```
