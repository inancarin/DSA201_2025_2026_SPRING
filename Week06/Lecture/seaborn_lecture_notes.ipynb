```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# let's calc avg total_bill for each day and display it

tips.groupby("day")["total_bill"].mean().plot(kind="bar")
```

```python
tips.head()
```

```python
sns.barplot(x=tips["day"], y=tips["total_bill"])
```

```python
# sns.barplot(x=tips["day"], y=tips["total_bill"])
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
sns.barplot(x="day", y="tip", data=tips, hue="sex")
```
