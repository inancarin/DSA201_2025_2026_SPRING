```python
for i in range(1000):
    res = titanic.sample(5, replace=True)
    if len(res.index.unique()) != 5:
        print(res.index)
```

```python
for i in range(1000):
    res = titanic.sample(5, replace=False)
    if len(res.index.unique()) != 5:
        print(res.index)
```

```python
titanic.head()
```

```python
n_iter = 1000
scores = []

conf_interval = 95


for i in range(n_iter):
    cur_avg = titanic["Fare"].sample(100).mean()
    scores.append(cur_avg)

scores.sort()
idx = int((n_iter * ((100 - conf_interval)/2)) / 100)

thr1 = scores[idx]
thr2 = scores[-idx]

# point1 = (thr1, 0)
# point2 = (thr1, 120)

sns.displot(scores)
plt.plot([thr1, thr1], [0, 120], color="red", linestyle="dashed", linewidth=3, label=format(thr1, ".2f"))
plt.plot([thr2, thr2], [0, 120], color="red", linestyle="dashed", linewidth=3, label=format(thr2, ".2f"))
plt.legend()
plt.show()
```
