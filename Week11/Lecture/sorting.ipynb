```python
def selectionSort(arr):
    for i in range(len(arr)):
        minIndex = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[minIndex]: # cost -> c1
                minIndex = j
        # cost of swap operation -> c2
        temp = arr[i]
        arr[i] = arr[minIndex]
        arr[minIndex] = temp
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr) # unsorted
selectionSort(myArr)
print("Sorted:", myArr) # sorted
```

```python
def bubbleSort(arr):
    for i in range(len(arr)):
        end = len(arr)-i
        for j in range(0, end-1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr) # unsorted
bubbleSort(myArr)
print("Sorted:", myArr) # sorted
```

```python
def mergeTwoSortedLists(arr1, arr2):
    result = []
    idx1, idx2 = 0, 0
    while idx1 < len(arr1) and idx2 < len(arr2):
        if arr1[idx1] < arr2[idx2]:
            result.append(arr1[idx1])
            idx1 += 1
        else:
            result.append(arr2[idx2])
            idx2 += 1
    
    while idx1 < len(arr1):
        result.append(arr1[idx1])
        idx1 += 1
    
    while idx2 < len(arr2):
        result.append(arr2[idx2])
        idx2 += 1
    
    return result
```

```python
res = mergeTwoSortedLists([10, 20, 30, 40], [5, 12, 22, 25])
res
```

```python
def mergeSort(arr): # T(n) = 2 * T(n/2) + O(n) = O(n logn)
    if len(arr) == 1:
        return arr
    left = arr[: len(arr)//2] # unsorted
    right = arr[len(arr)//2: ] # unsorted

    left_sorted = mergeSort(left) # T(n/2)
    right_sorted = mergeSort(right) # T(n/2)
    result = mergeTwoSortedLists(left_sorted, right_sorted) # O(n)
    return result
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr) # unsorted
myArr = mergeSort(myArr)
print("Sorted:", myArr) # sorted
```
