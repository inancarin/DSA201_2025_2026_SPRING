```python
def selectionSort(arr):
    for i in range(len(arr)):
        minIndex = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[minIndex]: # cost -> c1
                minIndex = j
        # cost of swap operation -> c2
        temp = arr[i]
        arr[i] = arr[minIndex]
        arr[minIndex] = temp
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr) # unsorted
selectionSort(myArr)
print("Sorted:", myArr) # sorted
```

```python
def bubbleSort(arr):
    for i in range(len(arr)):
        end = len(arr)-i
        for j in range(0, end-1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr) # unsorted
bubbleSort(myArr)
print("Sorted:", myArr) # sorted
```

```python
def mergeTwoSortedLists(arr1, arr2):
    result = []
    idx1, idx2 = 0, 0
    while idx1 < len(arr1) and idx2 < len(arr2):
        if arr1[idx1] < arr2[idx2]:
            result.append(arr1[idx1])
            idx1 += 1
        else:
            result.append(arr2[idx2])
            idx2 += 1
    
    while idx1 < len(arr1):
        result.append(arr1[idx1])
        idx1 += 1
    
    while idx2 < len(arr2):
        result.append(arr2[idx2])
        idx2 += 1
    
    return result
```

```python
def mergeSort(arr): # T(n) = 2 * T(n/2) + O(n) = O(n logn)
    if len(arr) == 1:
        return arr
    left = arr[: len(arr)//2] # unsorted
    right = arr[len(arr)//2: ] # unsorted

    left_sorted = mergeSort(left) # T(n/2)
    right_sorted = mergeSort(right) # T(n/2)
    result = mergeTwoSortedLists(left_sorted, right_sorted) # O(n)
    return result
```

```python
np.random.seed(42)
myArr = np.random.randint(1, 100, 10)
print("Unsorted:", myArr) # unsorted
myArr = mergeSort(myArr)
print("Sorted:", myArr) # sorted
```

```python
def naiveSolutionEx1(arr): # T(n) = O(n^2)
    if len(arr) < 2:
        return
    minDiff = abs(arr[1] - arr[0])
    num1, num2 = arr[0], arr[1]
    for i in range(len(arr)-1):
        for j in range(i+1, len(arr)):
            cur_diff = abs(arr[i] - arr[j])
            if cur_diff < minDiff:
                minDiff = cur_diff
                num1, num2 = arr[i], arr[j]

    print(num1, ",", num2) 
```

```python
myArr = [10, 8, 4, 12, 7, 25]
naiveSolutionEx1(myArr)
```

```python
def FindPairWithMinDiff(arr): # T(n) = O(n logn) + O(n) = O(n logn)
    if len(arr) < 2:
        return
    arr = sorted(arr) # O(n logn)
    minDiff = arr[1] - arr[0]
    num1, num2 = arr[0], arr[1]

    for i in range(1, len(arr)-1):
        curDiff = arr[i+1] - arr[i]
        if curDiff < minDiff:
            minDiff = curDiff
            num1, num2 = arr[i], arr[i+1]
    
    print(num1, ",", num2)
```

```python
myArr = [10, 8, 4, 12, 7, 25]
FindPairWithMinDiff(myArr)
```

```python
def findMissingNumber_v1(arr): # T(n) = O(n logn) + O(n) = O(n logn)
    arr = sorted(arr) # O(n logn)
    if arr[0] != 1:
        return 1
    for i in range(len(arr)-1): # O(n)
        if arr[i+1] != arr[i] + 1:
            return arr[i] + 1
        
    return len(arr) + 1
```

```python
arr = [2, 3, 1, 5]
findMissingNumber_v1(arr)
```

```python
def findMissingNumber_v2(arr): # T(n) = O(n)
    n = len(arr) + 1
    expected_sum = n * (n+1) / 2
    real_sum = sum(arr) # O(n)
    return expected_sum - real_sum

arr = [2, 3, 1, 5]
findMissingNumber_v2(arr)
```

```python
def findMissingNumber_v3(arr): # T(n) = O(n) + O(n) = O(n)
    n = len(arr) + 1
    mySet = set()
    for elem in arr: # O(n)
        mySet.add(elem)
    
    for i in range(1, n+1): # O(n)
        if i not in mySet:
            return i

arr = [2, 3, 1, 5]
findMissingNumber_v3(arr)
```

```python
def ex3(arr, target): # T(n) = O(n logn) + O(n) = O(n logn)
    if len(arr) < 2:
        return
    arr = sorted(arr) # O(n logn)

    # two pointer tech.
    i, j = 0, len(arr)-1
    minDiff = abs(target - (arr[j] + arr[i]))
    num1, num2 = arr[i], arr[j]

    while i < j: # O(n)
        cur_sum = arr[j] + arr[i]
        diff = abs(target - cur_sum)
        if diff < minDiff:
            minDiff = abs(target - (arr[j] + arr[i]))
            num1, num2 = arr[i], arr[j]
        if cur_sum < target:
            i += 1
        elif cur_sum > target:
            j -= 1
        else:
            break
    print(num1, ",", num2)
```

```python
myList = [22, 10, 28, 29, 30, 40]
target = 54
ex3(myList, target)
```
