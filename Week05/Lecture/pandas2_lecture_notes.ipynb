```python
def isEven(num):
    if num % 2 == 0:
        return True
    else:
        return False

myList = [10, 20, 21, 17, 24, 16, 15, 100, 104]

res = []
for elem in myList:
    if isEven(elem):
        res.append(elem)

print(res)
```

```python
# try to get the last name of each passenger - WRONG APPROACH -> return 0th row only
titanic["Name"].str.split(", ")[0][0]
```

```python
# try to get the last name of each passenger - CORRECT APPROACH

titanic["Name"].apply(lambda s:s.split(", ")[0])
```

```python
# let's find the passengers who has "bra" in their lastnames - NOT CORRECT

titanic[titanic["Name"].str.lower().str.contains("bra")]
```

```python
# let's find the passengers who has "bra" in their lastnames

titanic[titanic["Name"].apply(lambda s:"bra" in s.lower().split(", ")[0])]
```
