```python
myColorList = ["red", "blue", "green", "yellow", "orange"]

x_values = range(1, 6)
for k in range(1, 5):
    y_values = []
    for val in x_values:
        y_values.append(val * k)
    plt.plot(x_values, y_values, label="y=" + str(k) + "x", color=myColorList[k-1])

plt.legend()
plt.xticks(range(1, 6))
plt.yticks(range(1, 21, 2))
plt.show()
```
