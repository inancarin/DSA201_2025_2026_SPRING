```python
myColorList = ["red", "blue", "green", "yellow", "orange"]

x_values = range(1, 6)
for k in range(1, 5):
    y_values = []
    for val in x_values:
        y_values.append(val * k)
    plt.plot(x_values, y_values, label="y=" + str(k) + "x", color=myColorList[k-1])

plt.legend()
plt.xticks(range(1, 6))
plt.yticks(range(1, 21, 2))
plt.show()
```

```python
f = open("iris.txt")

lines = f.readlines()

f.close()

x_values, y_values, c_list = [], [], []
d = dict()
for line in lines[1:]:
    line = line.strip()
    info = line.split(",") # ALWAYS returns a list of strings
    sepal_l = float(info[1])
    sepal_w = float(info[2])
    species = info[-1]
    x_values.append(sepal_l)
    y_values.append(sepal_w)
    
    if species not in d:
        d[species] = len(d)
    
    c_list.append(d[species])

plt.scatter(x_values, y_values, c=c_list)
plt.xlabel("Sepal Length in cm")
plt.ylabel("Sepal Width in cm")
plt.show()
```

```python
f = open("iris.txt")

lines = f.readlines()

f.close()

d = dict()
for line in lines[1:]:
    line = line.strip()
    info = line.split(",") # ALWAYS returns a list of strings
    sepal_l = float(info[1])
    sepal_w = float(info[2])
    species = info[-1]

    if species not in d:
        d[species] = {"x": [], "y":[]}
    
    d[species]["x"].append(sepal_l)
    d[species]["y"].append(sepal_w)

for k, v in d.items():
    plt.scatter(v["x"], v["y"], label=k)

plt.legend()
plt.show()
```

```python
file = open("GOOGL.txt")

lines = file.readlines()

file.close()

d = {}

for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    year = int(info[0].split("-")[0])
    close = float(info[4])
    if year not in [2004, 2022]:
        if year not in d:
            d[year] = []
        d[year].append(close)

plt.figure(figsize=(15, 10))
for year, close_values in d.items():
    plt.plot(range(1, len(close_values)+1), close_values, label=year)

plt.legend()
plt.show()
```
