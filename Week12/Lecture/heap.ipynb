```python
import heapq
import numpy as np
import time
```

```python
def insert(heap, val): # assuming heap is a max-heap
    heap.append(val)
    i = len(heap) - 1
    parent_idx = (i - 1) // 2
    while parent_idx >= 0 and heap[parent_idx] < heap[i]:
        temp = heap[parent_idx]
        heap[parent_idx] = heap[i]
        heap[i] = temp
        i = parent_idx
        parent_idx = (i - 1) // 2
```

```python
myHeap = [88, 11, 22, 3, 5, 1, 19]
insert(myHeap, 25)
print(myHeap)
```

```python
def removeRoot(heap): # assume that heap is a maxheap
    if len(heap) == 0:
        return
    if len(heap) == 1:
        return heap.pop()
    temp = heap[0]
    heap[0] = heap[-1]
    heap[-1] = temp
    originalRoot = heap.pop()
    i = 0
    while 2 * i + 1 < len(heap):
        left = 2*i+1
        right = 2*i+2
        if heap[i] >= heap[left]:
            if right >= len(heap):
                break
            elif right < len(heap) and heap[i] >= heap[right]:
                break
        if heap[i] < heap[left]:
            greaterIdx = left
        if right < len(heap) and heap[right] > heap[left]:
            greaterIdx = right
        
        temp = heap[i]
        heap[i] = heap[greaterIdx]
        heap[greaterIdx] = temp
        i = greaterIdx
    return originalRoot
```

```python
myHeap = [59, 42, 44, 32, 39, 30, 13, 22, 29, 14, 33, 9, 17]
removedElem = removeRoot(myHeap)
print(removedElem)
print(myHeap)
```

```python
def heapSort(myList): # T(n) = O(n) + O(n logn) = O(n logn)
    res = []
    heapq.heapify(myList) # O(n)

    for i in range(len(myList)): # n
        minVal = heapq.heappop(myList) # logn
        res.append(minVal)
    
    return res
```

```python
myList = [10, 4, 2, 20, 100, 17, 18]
myList = heapSort(myList)
myList
```

```python
def find_kth_largest_v1(nums, k): # T(n) = O(n logn)
    nums = sorted(nums)
    return nums[-k]
```

```python
def find_kth_largest_v2(nums, k): # T(n) = O(n) + O(k logn)
    return heapq.nlargest(k, nums)[-1]
```

```python
def find_kth_largest_v3(nums, k): # T(n) = O(n logk)
    heap = []

    for val in nums: # n
        heapq.heappush(heap, val) # O(logk)
        if len(heap) == k+1:
            minVal = heapq.heappop(heap)
        
    return heap[0]
```

```python
def find_kth_largest_v4(nums, k): # T(n) = O(n) + O(k logn)
    heapq._heapify_max(nums) # convert into a max heap -> O(n)

    for i in range(k): # k
        maxVal = heapq._heappop_max(nums) # O(log n)
    
    return maxVal
```

```python
k = 10
n = 50000000

np.random.seed(42)
nums = np.random.randint(0, 1000000000, n).tolist()

start = time.time()
res = find_kth_largest_v1(nums, k)
end = time.time()

print("v1:", res, end-start)

np.random.seed(42)
nums = np.random.randint(0, 1000000000, n).tolist()

start = time.time()
res = find_kth_largest_v2(nums, k)
end = time.time()

print("v2:", res, end-start)

np.random.seed(42)
nums = np.random.randint(0, 1000000000, n).tolist()

start = time.time()
res = find_kth_largest_v3(nums, k)
end = time.time()

print("v3:", res, end-start)

np.random.seed(42)
nums = np.random.randint(0, 1000000000, n).tolist()

start = time.time()
res = find_kth_largest_v4(nums, k)
end = time.time()

print("v4:", res, end-start)
```

```python
def sortAlmostSortedArray(nums, k): # T(n) = O(n*2*logk) + O(k*logk) = O(n logk)
    result = []
    heap = []
    for val in nums: # n
        heapq.heappush(heap, val) # O(log k)
        if len(heap) == k+1:
            minVal = heapq.heappop(heap) # O(log k)
            result.append(minVal)
    
    while len(heap) > 0: # k
        minVal = heapq.heappop(heap) # O(log k)
        result.append(minVal)
    
    return result
```

```python
nums = [6, 5, 3, 2, 8, 10, 9]
k = 3
res = sortAlmostSortedArray(nums, k)
res
```

```python
def merge_k_sorted_lists(lists): # T(n) = O(k logk) + O(2*n*logk) = O(n logk)
    k = len(lists)
    heap = []
    for i in range(k): # k
        heapq.heappush(heap, (lists[i][0], i, 0)) # O(log k)
    
    res = []
    while len(heap) > 0: # n
        minVal, list_idx, elem_idx = heapq.heappop(heap) # O(log k)
        res.append(minVal)
        elem_idx += 1
        if elem_idx < len(lists[list_idx]):
            heapq.heappush(heap, (lists[list_idx][elem_idx], list_idx, elem_idx)) # O(log k)
    
    return res
```

```python
lists = [[1, 4, 5], [1, 3, 4], [2, 6]] 
merge_k_sorted_lists(lists)
```
