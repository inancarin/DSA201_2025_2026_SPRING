```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
class BST:
    def __init__(self):
        self.root = None
    
    def search(self, value): # worst-case -> O(n) / average-case -> O(logn)
        if self.root == None:
            return False
        cur = self.root
        while True:
            if cur.data == value:
                return True
            elif cur.data > value:
                cur = cur.left
                if cur == None:
                    return False
            else:
                cur = cur.right
                if cur == None:
                    return False

    def insert(self, data): # worst-case -> O(n) / average-case -> O(logn)
        newNode = Node(data)
        if self.root == None: # if the tree is empty
            self.root = newNode
        else:
            parent = self.root
            while True:
                if parent.data == data:
                    return # prevent having duplicated values in bst
                if parent.data > data:
                    if parent.left == None:
                        parent.left = newNode
                        break
                    parent = parent.left
                elif parent.data < data:
                    if parent.right == None:
                        parent.right = newNode
                        break
                    parent = parent.right
    
    def inOrderHelper(self, parent):
        if parent.left != None:
            self.inOrderHelper(parent.left)
        print(parent.data, end=" ")
        if parent.right != None:
            self.inOrderHelper(parent.right)
    
    def inOrder(self):
        if self.root == None:
            print("BST is empty!")
        else:
            self.inOrderHelper(self.root)
```

```python
bst = BST()
bst.insert(60)
bst.insert(55)
bst.insert(100)
bst.insert(45)
bst.insert(57)
bst.insert(67)
bst.insert(107)
```

```python
def findMin(root): # T(n) = O(n) worst-case, O(log n) in average
    if root == None:
        return None
    cur = root
    while cur.left != None:
        cur = cur.left
    
    return cur.data
```

```python
findMin(bst.root)
```

```python
def LCA(root, val1, val2): # T(n) = O(n) worst-case, O(log n) in average
    if root == None:
        return None
    cur = root
    while (cur.data > val1 and cur.data > val2) or (cur.data < val1 and cur.data < val2):
        if cur.data > val1 and cur.data > val2:
            cur = cur.left
        else:
            cur = cur.right
    return cur.data
```

```python
def helper(node, values):
    if node.left != None:
        helper(node.left, values)
    values.append(node.data)
    if node.right != None:
        helper(node.right, values)

def findKthSmallestNumber(root, k):
    if root == None:
        return None
    values = []
    helper(root, values)
    if k-1 < len(values):
        return values[k-1]
    else:
        print("K is too large!")
```

```python
def findAllAncestors(root, val):
    if root == None:
        return []
    result = []
    cur = root
    while cur.data != val:
        result.append(cur.data)
        if cur.data > val:
            cur = cur.left
        else:
            cur = cur.right
        
        if cur == None:
            print("there is no such value in bst")
            return
    result.append(cur.data)
    return result
```
