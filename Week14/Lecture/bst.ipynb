```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
class BST:
    def __init__(self):
        self.root = None
    
    def search(self, value): # worst-case -> O(n) / average-case -> O(logn)
        if self.root == None:
            return False
        cur = self.root
        while True:
            if cur.data == value:
                return True
            elif cur.data > value:
                cur = cur.left
                if cur == None:
                    return False
            else:
                cur = cur.right
                if cur == None:
                    return False

    def insert(self, data): # worst-case -> O(n) / average-case -> O(logn)
        newNode = Node(data)
        if self.root == None: # if the tree is empty
            self.root = newNode
        else:
            parent = self.root
            while True:
                if parent.data == data:
                    return # prevent having duplicated values in bst
                if parent.data > data:
                    if parent.left == None:
                        parent.left = newNode
                        break
                    parent = parent.left
                elif parent.data < data:
                    if parent.right == None:
                        parent.right = newNode
                        break
                    parent = parent.right
    
    def inOrderHelper(self, parent):
        if parent.left != None:
            self.inOrderHelper(parent.left)
        print(parent.data, end=" ")
        if parent.right != None:
            self.inOrderHelper(parent.right)
    
    def inOrder(self):
        if self.root == None:
            print("BST is empty!")
        else:
            self.inOrderHelper(self.root)
```

```python
bst = BST()
bst.insert(60)
bst.insert(55)
bst.insert(100)
bst.insert(45)
bst.insert(57)
bst.insert(67)
bst.insert(107)
```
