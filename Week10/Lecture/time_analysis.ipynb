```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
def linearSearch(myList, val):
    for elem in myList:
        if elem == val:
            return True
    
    return False
```

```python
numbers = [10, 5, 17, 22, 21, 100, 5]
val = 17
linearSearch(numbers, val)
```

```python
def binarySearch(myList, val):
    low, high = 0, len(myList)-1 # initial values

    while low <= high:
        mid = (low + high) // 2
        if myList[mid] == val:
            return True
        elif myList[mid] > val:
            high = mid - 1
        else:
            low = mid + 1
    return False
```

```python
def binarySearchRecursive(myList, val, low, high):
    if low > high:
        return False
    mid = (low + high) // 2
    if myList[mid] == val:
        return True
    elif myList[mid] > val:
        return binarySearchRecursive(myList, val, low, mid-1)
    else:
        return binarySearchRecursive(myList, val, mid+1, high)
```

```python
numbers = np.arange(1, 100, 2) # odd value
val = 17

#binarySearch(numbers, val)
binarySearchRecursive(numbers, val, 0, len(numbers)-1)
```

```python
#max_number = 32
for num in range(20, 30):
    max_number = 2 ** num

    numbers = np.arange(1, max_number, 2) # odd values (already sorted)
    val = 18

    start = time.time()
    res = linearSearch(numbers, val)
    end = time.time()

    diff = end - start
    print("Linear Search takes", diff, "seconds when max_numebr is", max_number)

    start = time.time()
    res = binarySearch(numbers, val)
    end = time.time()

    diff = end - start
    print("Binary Search takes", diff, "seconds when max_numebr is", max_number)
```

```python
# Ex1

np.random.seed(42)
numbers = np.random.randint(1, 1000000, 100000)

def findMax(numbers): # T(n) = O(1) + O(n) = O(n)
    if len(numbers) == 0: # O(1)
        return None
    maxValue = numbers[0]
    for idx in range(1, len(numbers)): # O(n-1) = O(n)
        if numbers[idx] > maxValue:
            maxValue = numbers[idx]
    
    return maxValue

findMax(numbers)
```

```python
x_values, y_values = [], []
for i in range(2, 25):
    n = 2 ** i
    np.random.seed(42)
    numbers = np.random.randint(1, 1000000, n)

    start = time.time()
    findMax(numbers)
    end = time.time()

    diff = end - start
    x_values.append(n)
    y_values.append(diff)

plt.plot(x_values, y_values)
plt.xlabel("Input size")
plt.ylabel("Time needed")
plt.show()
```

```python
# finding number of unique values using a list

np.random.seed(42)
numbers = np.random.randint(1, 1000000, 100000)

uniqueValues = []
for val in numbers:
    if val not in uniqueValues:
        uniqueValues.append(val)

print(len(uniqueValues))

# T(n) = O(0) + O(1) + O(2) + O(3) + ... + O(n-1) = O(n^2)
```

```python
# finding number of unique values using a set

np.random.seed(42)
numbers = np.random.randint(1, 1000000, 100000)

uniqueValues = set()
for val in numbers:
    if val not in uniqueValues:
        uniqueValues.add(val)

print(len(uniqueValues))

# T(n) = O(1) + O(1) + O(1) + ... + O(1) = O(n)
```

```python
"""
Fibonacci Series

0, 1, 1, 2, 3, 5, 8, 13, 21, ...

"""

def fibonacci(n): # T(n) = T(n-1) + T(n-2) + O(1) = O(2^n)
    if n == 0 or n == 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(40)
```

```python
def binarySearchRecursive(myList, val, low, high): # T(n) = O(1) + T(n/2) = O(log n)
    if low > high:
        return False
    mid = (low + high) // 2
    if myList[mid] == val:
        return True
    elif myList[mid] > val:
        return binarySearchRecursive(myList, val, low, mid-1)
    else:
        return binarySearchRecursive(myList, val, mid+1, high)
```

```python
"""
f0 -> 3
f1 -> 5
f2 -> 8
"""

def fibonacci_v2(n): # T(n) = O(1) + O(n-2) = O(n)
    f0, f1, f2 = 0, 1, 1

    for i in range(3, n+1):
        f0 = f1
        f1 = f2
        f2 = f0 + f1
    return f2

fibonacci_v2(40)
```
